# Train set data analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_train = pd.read_csv("X_train.csv")
```

```python
# Plot 1: Histogram of sequence lengths
plt.figure(figsize=(10, 6))
sns.histplot(data_train['length'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of Peptide Sequence Lengths', fontsize=16)
plt.xlabel('Sequence Length', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot 2: Bar plot of unique sequence lengths
unique_lengths = data_train['length'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
sns.barplot(x=unique_lengths.index, y=unique_lengths.values, palette="viridis")
plt.title('Frequency of Unique Peptide Sequence Lengths', fontsize=16)
plt.xlabel('Unique Sequence Length', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
